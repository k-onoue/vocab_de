```python
import gensim
model = gensim.models.KeyedVectors.load_word2vec_format('./../contents_pre-trained/model.vec',binary=False)
```

```python
word = '人生'
print(f'{word}の特徴量のベクトル')
print(model[word])
model[word].shape

print(f'{word}の類似語')
for item ,value in model.most_similar(word):
    print(item,value)
```
